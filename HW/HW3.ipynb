association rules
- extract association rules that have lift greater than 0.5, using itemsets from fpgrowth with support greater than 0.05

```python
def serv_rules(freq_itemsets,metrics,threshold):
    asso_rules = association_rules(freq_itemsets, metric=metrics, min_threshold=threshold)
    return asso_rules.sort_values(by='lift', ascending=False)[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
itemsets_great_5 = fpgrowth(serv_oneHot, min_support=0.05)
rules_fpgrowth = serv_rules(itemsets_great_5,"lift",0.5)
rules_fpgrowth
```

- extract association rules that have support greater than 0.0003, using itemsets from fpmax with support greater than 0.0001

```python
itemsets_great_0001 = fpmax(serv_oneHot, min_support = 0.0001)
rules_fpmax = association_rules(itemsets_great_0001, 'support', 0.0003, support_only = True)
rules_fpmax
```

make prediction (draw conclusion)
- if you were to make suggestions for DHS to consider increase the offering of one service, say, “Families_Receiving_Child_Welfare_Services” (S06), what other services would you suggest to offer together with? 
- run predict({“S06”}, rules, consequents_only=False), with rules generated from different thresholds and algorithms


```python
def predict(antecedent, rules, consequents_only = False):
    
    # get the rules for this antecedent
    preds = rules[rules['antecedents'] == antecedent]
    if consequents_only:
        # a way to convert a frozen set with one element to string
        preds = preds['consequents'].apply(iter).apply(next)
    return preds
```

```python
itemsets_fpgrowth1 = fpgrowth(serv_oneHot, min_support=0.05)
itemsets_fpgrowth2 = fpgrowth(serv_oneHot, min_support = 0.01)
itemsets_fpgrowth3 = fpgrowth(serv_oneHot, min_support = 0.1)
itemsets_fpgrowth4 = fpgrowth(serv_oneHot, min_support = 0.25)

itemsets_fpmax1 = fpmax(serv_oneHot, min_support = 0.001)
itemsets_fpmax2 = fpmax(serv_oneHot, min_support = 0.005)
itemsets_fpmax3 = fpmax(serv_oneHot, min_support = 0.0001)
itemsets_fpmax4 = fpmax(serv_oneHot, min_support = 0.0025)

itemsets_fpgrowth = [itemsets_fpgrowth1, itemsets_fpgrowth2, itemsets_fpgrowth3, itemsets_fpgrowth4]
itemsets_fpmax = [itemsets_fpmax1, itemsets_fpmax2, itemsets_fpmax3, itemsets_fpmax4]
```

```python
predictions = []
for itemset in itemsets_fpgrowth:
    rules = serv_rules(itemset, 'confidence', 0.5)
    prediction = predict({'S06'}, rules, consequents_only = False)
    predictions.append(prediction)
    
```
